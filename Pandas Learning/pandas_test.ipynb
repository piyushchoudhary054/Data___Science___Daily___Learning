```python
# SECTION E: GROUPBY OPERATIONS
# 1. Find the average score for each performance category.
df["total_score"] = (
    df['ai_dependency_score'] +
    df['ai_generated_content_percentage'] +
    df['ai_prompts_per_week'] +
    df['ai_ethics_score'] +
    df['last_exam_score'] +
    df['assignment_scores_avg'] +
    df['concept_understanding_score']
)

df.groupby('performance_category')['total_score'].mean()
```
