```python
#Write a program to check if a given year is a leap year. A leap year is divisible by 4 but not by 100
#unless it is also divisible by 400.

year = int(input("enter the year"))

if(year%4 == 0 and year%100 !=0 or year%400 == 0):
    print("leap year")
else:
    print("not leap year")
        
```

```python
'''Write a basic calculator program that takes two numbers and an operator (+, -, *, /) as input and
performs the specified operation. Print the result based on the operation.'''

num1 = int(input("enter the first number"))
num2 = int(input("enter the second number"))
operator = str(input("enter the operator"))

if(operator == '+'):
    print(num1+num2)
elif(operator == '-'):    
    print(num1-num2)
elif(operator == '*'):
    print(num1*num2)
elif(operator == '/'):
    print(num1/num2)
else:
    print("invalid operator")    
```

```python
'''Write a program that calculates the Body Mass Index (BMI) based on user input for weight (in
kilograms) and height (in meters). Then categorize the BMI into:
Underweight (BMI < 18.5)
Normal weight (18.5 <= BMI < 24.9)
Overweight (25 <= BMI < 29.9)
Obesity (BMI >= 30)
Use the formula: BMI = weight / (height ** 2)'''

weight = int(input("enter the weight"))
height = float(input("enter the height:"))
BMI = weight/(height **2)

if(BMI>=30):
    print("Obesity")
elif(26<=BMI<29.9):
    print("Overweight")
elif(18.5<=BMI<24.9):
    print("Normal weight") 
elif(BMI<18.5):
    print("Underweight ")      
```

```python
'''Write a program that calculates the discount for a product based on its price:
If price is greater than 1000, discount is 10%
If price is between 500 and 1000, discount is 5%
Otherwise, no discount
Print the final price after applying the discount.'''
price = int(input("enter the price"))
if(price>=1000):
    discount = price*0.1
    print(f"the  price after discount is {price-discount}")
elif(500<=price<1000):
    discount = price*0.05
    print(f"the   after discount is {price-discount}")
else:
    print("no discount the price is",price)    
```
