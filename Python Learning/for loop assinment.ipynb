```python
'''16.Write a program to print all Prime numbers between 1 to n.'''
num = int(input("enter the number"))

for i in range(2,num+1):
    isprime = 1
    for j in range(2,i):
        if(i%j == 0):
            isprime = 0
            break
    if(isprime == 1):
        print(i)    
```
